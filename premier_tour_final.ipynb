## Datos informe casos de corrupción en Colombia



```python
bd_corr = pd.read_csv('data/base-de-datos-general_2.csv', sep=';')
bd_corr.head()
```

```python
bd_corr.columns
```

```python
bd_corr.dtypes
```

```python
bd_corr = bd_corr.rename(columns={'Tipo de corrupción':'tipo_corrupcion'})
```

```python
bd_corr.tipo_corrupcion.unique()
```

```python
df_cor_depto = bd_corr.groupby(['Departamento'])['tipo_corrupcion'].count().reset_index(name='count')
df_cor_depto
```

```python
import plotly.io as pio
pio.templates
```

```python
fig = px.bar(df_cor_depto, x='Departamento', y='count',
            color='count',
            template='plotly_dark',
            labels={'count':'Cantidad de casos'},
            title='Cantidad de casos de corrupción por departamento',
            height=400)
fig.show()
```

```python
bd_corr.Región.unique()
```

```python
# Crear máscara con una sola categoría de una columna
act_corr_reg_caribe = bd_corr[bd_corr.Región == 'Región Caribe']
act_corr_reg_caribe.head()
```

```python
act_corr_reg_caribe_xMun = act_corr_reg_caribe.groupby('Municipio')['tipo_corrupcion'].count().sort_values(ascending=False)
act_corr_reg_caribe_xMun.head(15)
```

```python
# Mapas
map_corr_deptos = bd_corr.groupby(['Departamento', 'Dep_Lat', 'Dep_Lng'])['tipo_corrupcion'].count().reset_index()
map_corr_deptos.head(32)
```

```python
map_corr_deptos.dtypes
```

```python
#map_corr_deptos.Dep_Lat = map_corr_deptos.Dep_Lat.astype(float)
map_corr_deptos.to_csv("map_corr_deptos_3.csv")
```

```python
map_corr_deptos.Departamento = map_corr_deptos.Departamento.replace({'BOGOTÁ, DISTRITO CAPITAL' : 'BOGOTA D.C.'})
map_corr_deptos.head()
```

```python
map_corr_deptos_4 = pd.read_csv('data/map_corr_deptos_4.csv', sep=';')
map_corr_deptos_4.head()
```

```python
map_corr_deptos_4.dtypes
```

```python
map_corr_deptos_4.Departamento = map_corr_deptos_4.Departamento.replace({'BOGOTÁ, DISTRITO CAPITAL' : 'BOGOTA D.C.'})
map_corr_deptos_4.Departamento = map_corr_deptos_4.Departamento.replace({'GUAJIRA' : 'LA GUAJIRA'})
map_corr_deptos_4.Departamento = map_corr_deptos_4.Departamento.replace({'NORTE SANTANDER' : 'NORTE DE SAN'})
map_corr_deptos_4.Departamento = map_corr_deptos_4.Departamento.replace({'SAN ANDRES, PROV.' : 'SAN ANDRES'})
map_corr_deptos_4.head()
```

```python
px.scatter_mapbox(map_corr_deptos_4,
                  lat='Dep_Lat',
                  lon='Dep_Lng',
                  color='Departamento',
                  size='tipo_corrupcion',
                  color_continuous_scale=px.colors.cyclical.IceFire, 
                  #size_max=5, 
                  zoom=4,
                  center = dict(
                    lat = 4.570868,
                    lon = -74.297333
                  ),
                  height= 600
)
```

```python
fig = go.Figure(go.Scattermapbox(
    lon = map_corr_deptos_4.Dep_Lng,
    lat = map_corr_deptos_4.Dep_Lat,
    mode = 'markers+text',
    marker = go.scattermapbox.Marker(size=map_corr_deptos_4.tipo_corrupcion,
                                    color = map_corr_deptos_4.tipo_corrupcion,
                                    #colorscale = 'Edge',
                                    showscale = True,
                                    sizemode = 'area', 
                                    opacity = 0.8
                                     ),
    hoverinfo = 'text',
    hovertext = 
    '<b>Departamento</b>: '+ map_corr_deptos_4['Departamento'].astype(str) + '<b>' +
    '<b>Cantidad de actos</b>: '+ map_corr_deptos_4['tipo_corrupcion'].astype(str) + '<b>'
))

fig.update_layout(
    hovermode = 'x',
    margin = dict(r=0, l=0, b=0, t=0),
    mapbox = dict(
        accesstoken = token_map_plot,
        style = 'dark',
        zoom = 4.5,
        center = dict(
            lat = 4.570868,
            lon = -74.297333
        )
    ),
    showlegend = True,
    autosize = True
)
```
